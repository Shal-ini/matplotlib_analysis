```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import seaborn as sns
df = pd.DataFrame(np.random.randn(100,4),index=pd.date_range('9/3/2022', periods=100),columns=list('ABCD'))
df = df.cumsum()
sns.relplot(x=df.index, y="A", kind="line", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
df=pd.DataFrame(np.random.randn(100,4),index=pd.date_range("9/3/2022", periods=100),columns=list("ABCD"))
df=df.cumsum()
sns.lineplot(x=df.index, y='A', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
#when you use two variables for plotting, it is known as a bivariate plot.
#Here is a simple example
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
tips = sns.load_dataset("tips")
sns.displot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
#you can add color to this example as follows
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
tips = sns.load_dataset("tips")
sns.displot(x="total_bill", y="tip", hue="size", data=tips)
plt.show()
```

```python
#you can also customize the size of bins and add ticks on the x- y-axes
#rug plot
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
tips = sns.load_dataset("tips")
sns.displot(x="total_bill", y="tip", hue="size", data=tips, rug=True, kind='hist', bins=30)
plt.show()
```

```python
# A more interesting type of visualization is a bivariate KDE plot. 
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
tips = sns.load_dataset("tips")
sns.displot(x="total_bill", y="tip",data=tips,kind='kde')
plt.show()
```

```python
#you can add a rug plot to the output as follows:
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
tips = sns.load_dataset("tips")
sns.displot(x="total_bill", y="tip",data=tips, rug=True, kind='kde')
plt.show()
```

```python
#Based on columns in the dataframe , you can create individual visualizations 
#arranged in rows or columns. let's create a visualization based on the size of tips as foolows:
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
tips = sns.load_dataset("tips")
sns.displot(x="total_bill", y="tip", col="size", data=tips, rug=True, kind='kde')
plt.show()
```

```python
#You can arrange th individual graphs in rows as follows:
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
tips = sns.load_dataset("tips")
sns.displot(x="total_bill", y="tip",data=tips, rug=True, kind='kde', row="size")
plt.show()
```
